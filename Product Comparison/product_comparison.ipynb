```python
import pandas as pd

# Membaca file CSV ke dalam DataFrame
df = pd.read_csv("FINAL_BENERAN_YA ALLAH_AAMIIN.csv")

# Membuat kolom baru "Cluster" dengan meng-copy nilai dari kolom "Shade"


# Mengubah isi kolom "Cluster" yang awalnya "Cluster 1" - "Cluster 12" menjadi integer 1 - 12
df["Cluster"] = df["Cluster"].map(lambda x: int(x.split()[1]) if x.startswith("Cluster") else x)
df = df.rename(columns={"Cluster": "Shade"})
# Menyimpan DataFrame yang telah diubah
df.to_csv("FINAL_BENERAN_YA ALLAH_AAMIIN_updated.csv", index=False)

# Menampilkan DataFrame setelah diubah
df.head() 
```
