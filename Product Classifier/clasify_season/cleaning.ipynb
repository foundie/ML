```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("combined.csv")

df
```

```python
def process_value(value):
    # Menghapus tanda petik
    value = value.replace("'", "")
    # Mengubah nilai menjadi float
    value = tuple(float(v) for v in value.strip('()').split(','))
    return value

# Menggunakan fungsi apply pada kolom 'S1 Closest Color'
df['S1 Closest Color'] = df['S1 Closest Color'].apply(process_value)
df['S1 Closest Color'] = df['S1 Closest Color'].apply(list)
df
```

```python
# Menggunakan fungsi apply pada kolom 'S1 Closest Color'
df['S2 Closest Color'] = df['S2 Closest Color'].apply(process_value)
df['S2 Closest Color'] = df['S2 Closest Color'].apply(list)
df
```

```python
df_cleaned=df.dropna()
df_cleaned
```

```python
df_cleaned.loc[:, 'Season 1 Name'] = df_cleaned['Season 1 Name'].str.strip("'")
df_cleaned.loc[:, 'Season 2 Name'] = df_cleaned['Season 2 Name'].str.strip("'")

df_cleaned
```

```python
df_cleaned.to_excel("df_cleaned_v1.xlsx", index=False)
```
